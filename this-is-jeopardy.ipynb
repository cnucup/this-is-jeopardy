```python
def partial_df(df, word_list):
    
    #partial_df = pd.DataFrame()
    
    filter = lambda x: all(word.lower() in x.lower() for word in word_list)
    
    #partial_df = df.loc[df["Question"].apply(filter)]
    
    #return partial_df
    return df.loc[df["Question"].apply(filter)]
```

```python
display(partial_df(["King", "England"]))
partial_df = partial_df(["King", "England"])
display(partial_df.head(50))
```

### Here, we are cleaning the Value column and converting it into floating numbers to find the average values

```python
def value_converted_df (partial_df):
    partial_df.drop(partial_df[partial_df.Value == "None"].index, inplace=True)
#display(partial_df.head(40))
    partial_df["New_Value"] = pd.to_numeric(partial_df["Value"].str.strip("$").replace("[\,,]", "", regex=True), downcast='float')
    return partial_df
#display(partial_df.head(20))
#partial_df.dtypes
```

```python
#print(partial_df(["King"]))
#print(type(partial_df(["King"])))
value_converted_df(partial_df(["King"]))
#print(value_converted_df(partial_df(["King"])).New_Value.mean())
ave_value_king = value_converted_df(partial_df(["King"])).New_Value.mean()
print(ave_value_king)
```

### We are looking for unique answers and find the frequencies

```python
def find_unique_answers(dataset, word_list):
    new_data = partial_df(word_list).Answer
    (unique, counts) = np.unique(new_data, return_counts=True)
    frequencies = np.asarray((unique, counts)).T
    return frequencies
```

# Additional Stuff

question() class generates a question and answer. And then, we ask the question to the contestant, get the answer and check it. This game could have been written in a way it keeps asking questions as contestant want new question. A few contestants may have been added as well and we could keep points like a real contest.

```python
import numpy as np

class question():
    
    def __init__(self, dataset):
        i = np.random.choice(dataset.index)
        self.i = i
        self.dataset = dataset
        #print(i)
  
    def question_generator(self):
        return self.dataset["Question"][self.i]

    def question_answer(self):
        return self.dataset["Answer"][self.i]
```

```python
answer = input ("You want to test yourself? : Y/N")
if answer.lower() == "y":
    question1 = question(df)
    print(question1.question_generator())
    answer = input ("Please enter your answer : ")
    if answer.lower() == question1.question_answer().lower():
        print("Congratulations !")
    else:
        print("Wrong answer...")
        print(question1.question_answer())
```
